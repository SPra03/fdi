```python
from datapackage import Package, Resource
import pandas as pd
```

```python
# Required fields
required_fields_ = ["Subject_ID","Species","Strain","Animal_origin","Age","Weight","Sex","Group","Laboratory","StudyLeader","Published","Exclusion_in_origin_study","Exclusion_reason","Cause_of_Death","Injury_type","Injury_device","Injury_level","Injury_details"]
# getting the data_dict
data_dictionary = pd.read_csv("odc-sci_824/odc-sci_824_dictionary.csv",  encoding='utf-8', engine='python')
```

```python
# extracting the schema details and creating a field data structure out of it
fields = []

# iterate over each row in the dataframe
for index, row in data_dictionary.iterrows():
    # create a dictionary for the row
    row_data = {}
    # iterate over each column in the row
    for col_name in data_dictionary.columns:
        # add the column name and value to the dictionary
        if pd.isna(row[col_name]):
          # if the value is empty, fill it with None
          row_data[col_name] = None
          # print(col_name)
        else:
          row_data[col_name] = row[col_name]
    if row['VariableName'] in required_fields_:
      row_data['Required'] = True
    else:
      row_data['Required'] = False
    # append the row dictionary to the list of data
    fields.append(row_data)
```

# Creating a Data Package Folder

To create a data package, we need to create a folder with the same structure we want the data package object to have. Let's illustrate this with an example:

### Example: odc-sci_824_datapackage

Here, we will create a folder named `odc-sci_824_datapackage` with the following structure:

- odc-sci_824_datapackage
    - data
        - odc-sci_824.csv
        - odc-sci_824_dictionary.csv
    - readme.md
    - datapackage.json

In this structure:
- The `data` folder contains the actual data files: `odc-sci_824.csv` and `odc-sci_824_dictionary.csv`.
- The `readme.md` file is just a dummy example file.
- The `datapackage.json` file, which will be generated and saved at the end, holds the metadata for the data package.

This organized structure helps in packaging and sharing data in a standardized manner.

```python
import zipfile
import os

def zip_folder(folder_path, output_path):
    # Ensure the output path includes the .zip extension
    if not output_path.endswith('.zip'):
        output_path += '.zip'

    # Create a ZipFile object in write mode
    with zipfile.ZipFile(output_path, 'w', zipfile.ZIP_DEFLATED) as zipf:
        # Walk through all files and subdirectories in the folder
        for foldername, subfolders, filenames in os.walk(folder_path):
            for filename in filenames:
                file_path = os.path.join(foldername, filename)
                # Add each file to the archive with its relative path
                arcname = os.path.relpath(file_path, folder_path)
                zipf.write(file_path, arcname=arcname)

```

```python
def get_abstract(input_string):

  if len(input_string) <1:
    return {
      "STUDY PURPOSE": "",
      "DATA COLLECTED": ""
    }
  else:
    # position of "STUDY PURPOSE:" and "DATA COLLECTED:"
    study_purpose_start = input_string.find("STUDY PURPOSE:") + len("[STUDY PURPOSE:")
    data_collected_start = input_string.find("DATA COLLECTED:") + len("[DATA COLLECTED:")

    # Extract the text between "[STUDY PURPOSE:" and "[DATA COLLECTED:"
    study_purpose = input_string[study_purpose_start:data_collected_start - len("DATA COLLECTED:")-1]

    # Extract the text after "[DATA COLLECTED:"
    data_collected = input_string[data_collected_start:]

    # Print the extracted values
    return {
        "STUDY PURPOSE": study_purpose,
        "DATA COLLECTED": data_collected
    }

try:
  abstract = data['data']['attributes']['descriptions'][0]['description']
except:
  abstract = ""


get_abstract(abstract)
```
